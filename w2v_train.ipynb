```python
import jieba
import logging
from gensim.models import word2vec
```

```python
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)

# jieba custom setting.
jieba.set_dictionary('jieba_dict/dict.txt.big')

# load stopwords set
stopword_set = set()
with open('jieba_dict/stopwords.txt','r', encoding='utf-8') as stopwords:
    for stopword in stopwords:
        stopword_set.add(stopword.strip(' \n'))
```

```python
output = open('wiki_seg.txt', 'w', encoding='utf-8')
with open('wiki_zh_tw.txt', 'r', encoding='utf-8') as content :
    for texts_num, line in enumerate(content):
        line = line.strip('\n')
        words = jieba.cut(line, cut_all=False)
        for word in words:
            if word not in stopword_set:
                output.write(word.strip(' ') + ' ')
        output.write('\n')
        
        if (texts_num + 1) % 10000 == 0:
            logging.info("已完成前 %d 行的斷詞" % (texts_num + 1))
output.close()
```

## w2v

```python
# Set up logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)

# Settings
train_data = word2vec.LineSentence('wiki_seg.txt')
w2v_model_sg = word2vec.Word2Vec(
    train_data,
    min_count=5,
    size=250,
    workers=3,
    iter=10,
    window=5,
    sg=1,
    negative=5,
)

w2v_model_sg.save('w2v_sg.model')
```

```python
# test model
model = word2vec.Word2Vec.load('w2v_sg.model')
```
