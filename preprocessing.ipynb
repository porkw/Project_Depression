```python
from ckiptagger import WS, POS, NER
ws=WS("ckip")
pos=POS("ckip")
ner=NER("ckip")
```

```python
def about_ner(text):
    
    if str(text)=='nan':
        return ''
    
    ws_results=ws([text])
    pos_results=pos(ws_results)
    
    #print(ws_results)
    #print(pos_results)
    
    res=[]
    for i in zip(ws_results[0], pos_results[0]):
        #if (i[1]=='Na' or i[1]=='Nb' or i[1]=='Nc' or i[1]=='ORG' or i[1]=='GPE' or i[1]=='FAC' or\
        #    i[1]=='NORP' or i[1]=='PERSON' or i[1]=='VC'):
            res.append(i[0]+"-"+i[1])
    return res
```

```python
#text_parser
df['pos']=df['Content'].apply(lambda x: ','.join(about_ner(x)))
```

```python
#sentence_analysis
word_pos=[]
for terms in df[df['depression']=='1']['pos'].apply(lambda x: x.split(',')):
    for term in terms:
        word_pos.append(term)
```

```python
from collections import Counter
counts = Counter(word_pos)
#print(counts)
```

```python
#找最多字頻
for key, value in counts.most_common():
    print(key, value)
```
